# Propability ratio encoding

We first calculate the prop() of the target class , then the counter of it (1 - prop() ) then divide the prop() of target with counter prop() to get propabiblity ratio . 

```python
l1 = [0,1]
l2 = ['CA' , 'US' , 'PAK' , 'AFG' , 'IND' , 'UAE' , 'UK' , 'AUS']
dic = {'Survived':np.random.choice(l1 , size = 30) , 'Class' : np.random.choice(l2 , size = 30)}
df = pd.DataFrame(dic)
df.head()
```

```python
df['Class_1'] = df['Class'].astype(str).str[0]
df.sample(5)
```

```python

prob_df = df.groupby(['Class_1'])['Survived'].count()
prob_df = prob_df.to_frame()
prob_df
```

```python

prob_df['prop'] = prob_df['Survived']/np.sum(prob_df['Survived'])
prob_df
```

```python
prob_df = df.groupby(['Class_1'])['Survived'].mean()
prob_df
```

```python

prob_df = pd.DataFrame(prob_df)
prob_df
```

```python
prob_df['Died'] = 1 - prob_df['Survived']
prob_df
```

```python
prob_df['Propability_Ratio'] = prob_df['Survived']/prob_df['Died']
prob_df
```

```python
var = prob_df['Propability_Ratio'].to_dict()
var
```

```python
df['Class_1'] = df['Class_1'].map(var)
df.head(10)
# Nowthe final 'Class_1' columns will be used in ML regardless of all other columns 
```
